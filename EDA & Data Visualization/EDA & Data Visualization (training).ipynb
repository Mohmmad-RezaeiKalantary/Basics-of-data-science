# Cleaning Data

```python
columns_to_replace_nan = [
    'total_cases',
    'new_cases',
    'new_cases_smoothed',
    'total_deaths',
    'new_deaths',
    'new_deaths_smoothed',
    'total_cases_per_million',
    'new_cases_per_million',
    'new_cases_smoothed_per_million',
    'total_deaths_per_million',
    'new_deaths_per_million',
    'new_deaths_smoothed_per_million',
    'reproduction_rate',
    'icu_patients',
    'icu_patients_per_million',
    'hosp_patients',
    'hosp_patients_per_million',
    'weekly_icu_admissions',
    'weekly_icu_admissions_per_million',
    'weekly_hosp_admissions',
    'weekly_hosp_admissions_per_million',
    'total_tests',
    'new_tests',
    'total_tests_per_thousand',
    'new_tests_per_thousand',
    'new_tests_smoothed',
    'new_tests_smoothed_per_thousand',
    'positive_rate',
    'tests_per_case',
    'total_vaccinations',
    'people_vaccinated',
    'people_fully_vaccinated',
    'total_boosters',
    'new_vaccinations',
    'new_vaccinations_smoothed',
    'total_vaccinations_per_hundred',
    'people_vaccinated_per_hundred',
    'people_fully_vaccinated_per_hundred',
    'total_boosters_per_hundred',
    'new_vaccinations_smoothed_per_million',
    'new_people_vaccinated_smoothed',
    'new_people_vaccinated_smoothed_per_hundred',
    'excess_mortality_cumulative_absolute',
    'excess_mortality_cumulative',
    'excess_mortality',
    'excess_mortality_cumulative_per_million'
]

# Replace NaN values with 0 for the specified columns
data[columns_to_replace_nan] = data[columns_to_replace_nan].fillna(0)
```

```python
data.isnull().sum()
```

```python
# Display unique values in the 'tests_units' column
unique_tests_units = data['tests_units'].unique()

# Print the unique values
print(unique_tests_units)
```

```python
# Replace NaN values in the 'tests_units' column with "no test"
data['tests_units'] = data['tests_units'].fillna("no test")
```

```python
columns_to_flag_missing = [
    'stringency_index',
    'population_density',
    'median_age',
    'aged_65_older',
    'aged_70_older',
    'gdp_per_capita',
    'extreme_poverty',
    'cardiovasc_death_rate',
    'diabetes_prevalence',
    'female_smokers',
    'male_smokers',
    'handwashing_facilities',
    'hospital_beds_per_thousand',
    'life_expectancy',
    'human_development_index'
]

# Create binary indicator columns for missing values in specified columns
for column in columns_to_flag_missing:
    data[f'{column}_missing'] = data[column].isnull().astype(int)
```

```python
data.head(15)
```

```python
# Filter rows where 'stringency_index_missing' is 1 (indicating missing values)
countries_with_missing_stringency = data.loc[data['aged_65_older_missing'] == 1, 'location'].unique()

# Print the countries with missing values in 'stringency_index'
print("Countries with missing values in 'aged_65_older_missing':")
print(countries_with_missing_stringency)
```

```python
columns_to_check = [
    'stringency_index',
    'population_density',
    'median_age',
    'aged_65_older',
    'aged_70_older',
    'gdp_per_capita',
    'extreme_poverty',
    'cardiovasc_death_rate',
    'diabetes_prevalence',
    'female_smokers',
    'male_smokers',
    'handwashing_facilities',
    'hospital_beds_per_thousand',
    'life_expectancy',
    'human_development_index'
]

for column in columns_to_check:
    # Create a boolean mask for rows where the column's missing indicator is 1
    missing_indicator_column = f'{column}_missing'
    missing_data_mask = data[missing_indicator_column] == 1

    # Extract the unique countries from the filtered rows
    countries_with_missing_data = data.loc[missing_data_mask, 'location'].unique()

    # Print the countries with missing values in the current column
    print(f"Countries with missing values in '{column}':")
    print(countries_with_missing_data)
    print()
```

```python
data['location'].unique()
```

```python
# List of values to be removed from the main data
values_to_remove = ['World', 'Asia', 'Europe', 'North America', 'South America','European Union' , 'High income', 'Lower middle income', 'Low income', 'Upper middle income']

# Filter rows with specified values in the 'location' column
filtered_data = data[data['location'].isin(values_to_remove)]

# Create a separate dataframe with the filtered data
separate_dataframe = filtered_data.copy()

# Remove rows with specified values from the main data
data = data[~data['location'].isin(values_to_remove)]
```

# Advanced DA

```python
columns_to_check = [
    'stringency_index',
    'population_density',
    'median_age',
    'aged_65_older',
    'aged_70_older',
    'gdp_per_capita',
    'extreme_poverty',
    'cardiovasc_death_rate',
    'diabetes_prevalence',
    'female_smokers',
    'male_smokers',
    'handwashing_facilities',
    'hospital_beds_per_thousand',
    'life_expectancy',
    'human_development_index'
]

for column in columns_to_check:
    # Create a boolean mask for rows where the column's missing indicator is 1
    missing_indicator_column = f'{column}_missing'
    missing_data_mask = data[missing_indicator_column] == 1

    # Group by 'location' and sum the missing values for each country
    missing_values_by_country = data.loc[missing_data_mask, ['location', missing_indicator_column]].groupby('location').sum()

    # Print the countries with the number of missing values in the current column
    print(f"Number of missing values in '{column}' by country:")
    print(missing_values_by_country)
    print()
```

```python
import folium
from folium.plugins import MarkerCluster

# Find the row with the maximum total cases
max_total_cases_row = data.loc[data['total_cases'].idxmax()]

# Create a Folium map centered around the location with maximum total cases
map_center = [max_total_cases_row['total_cases'], max_total_cases_row['total_cases']]
my_map = folium.Map(location=map_center, zoom_start=2)

# Add markers for each location with total cases
marker_cluster = MarkerCluster().add_to(my_map)

for index, row in data.iterrows():
    popup_text = f"{row['location']} - Total Cases: {row['total_cases']}"
    marker_color = 'red' if row['location'] == max_total_cases_row['location'] else 'blue'

    folium.Marker([row['total_cases'], row['total_cases']],
                  popup=popup_text,
                  icon=folium.Icon(color=marker_color)
                 ).add_to(marker_cluster)

# Save the map as an HTML file or display it
my_map.save('covid_total_cases_map_max.html')
```

```python
import plotly.express as px

fig = px.scatter_geo(data,
                     locations="location",
                     locationmode="country names",
                     color="total_cases",
                     size="total_cases",
                     projection="natural earth",
                     title="Total COVID-19 Cases by Location",
                     hover_name="location",
                     size_max=50)

# Save the plot as an HTML file
fig.write_html('covid_total_cases_plot.html')
```
