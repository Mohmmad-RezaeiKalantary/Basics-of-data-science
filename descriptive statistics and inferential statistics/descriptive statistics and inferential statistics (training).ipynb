# descriptive statistics and inferential statistics

1.Are there significant trends in song popularity (Popularity) over time?

```python
# Group the data by release year and calculate the mean popularity for each year
popularity_by_year = data.groupby('Release Year')['Popularity'].mean().reset_index()

# Create a line plot to visualize popularity over time
plt.figure(figsize=(12, 6))
plt.plot(popularity_by_year['Release Year'], popularity_by_year['Popularity'], marker='o', linestyle='-', color='b')
plt.xlabel('Release Year')
plt.ylabel('Average Popularity')
plt.title('Popularity Trends Over Time')
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm

# Fit a linear regression model to test the trend in popularity over time
X = sm.add_constant(popularity_by_year['Release Year'])
Y = popularity_by_year['Popularity']
model = sm.OLS(Y, X).fit()
slope_pvalue = model.pvalues[1]

if slope_pvalue < 0.05:
    print("There is a significant trend in song popularity over time.")
else:
    print("There is no significant trend in song popularity over time.")

confidence_interval = model.conf_int().iloc[1]  # Confidence interval for the slope
print("95% Confidence Interval for the Slope:", confidence_interval)
```

4.Are certain geographic regions (counties) associated with higher accident severity?

```python
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import MultiComparison
from scipy.stats import kruskal

# Calculate the mean severity for each county
county_severity = data.groupby('County')['Severity'].mean().reset_index()

# Print the counties with the highest and lowest average severity
highest_severity_county = county_severity.loc[county_severity['Severity'].idxmax()]
lowest_severity_county = county_severity.loc[county_severity['Severity'].idxmin()]

print("County with the Highest Average Severity:")
print(highest_severity_county)

print("County with the Lowest Average Severity:")
print(lowest_severity_county)

# Assuming your dataset is named 'data' and contains the 'County' and 'Severity' columns
county_severity = data[['County', 'Severity']]

# Perform Kruskal-Wallis test
kruskal_result = kruskal(*[group['Severity'] for name, group in county_severity.groupby('County')])

# Determine if there are statistically significant differences
if kruskal_result.pvalue < 0.05:
    print("Certain geographic regions (counties) are associated with higher accident severity.")
else:
    print("There are no statistically significant differences in accident severity between counties.")
```

5.Is there a Relationship Between Visibility and Accident Severity?

```python
from scipy.stats import ttest_ind

# Assuming your dataset is named 'data' and contains the 'Visibility(mi)' and 'Severity' columns
visibility_severity = data[['Visibility(mi)', 'Severity']]

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Visibility(mi)', y='Severity', data=visibility_severity)
plt.xlabel('Visibility (mi)')
plt.ylabel('Severity')
plt.title('Relationship Between Visibility and Accident Severity')
plt.show()

# Perform a t-test to compare the means of severity between different visibility conditions
severe_visibility = visibility_severity[visibility_severity['Severity'] == 4]
moderate_visibility = visibility_severity[visibility_severity['Severity'] == 3]
mild_visibility = visibility_severity[visibility_severity['Severity'] == 2]
low_visibility = visibility_severity[visibility_severity['Severity'] == 1]

# Perform t-tests for each pair
t_stat_severe_moderate, p_value_severe_moderate = ttest_ind(severe_visibility['Visibility(mi)'], moderate_visibility['Visibility(mi)'])
t_stat_severe_mild, p_value_severe_mild = ttest_ind(severe_visibility['Visibility(mi)'], mild_visibility['Visibility(mi)'])
t_stat_severe_low, p_value_severe_low = ttest_ind(severe_visibility['Visibility(mi)'], low_visibility['Visibility(mi)'])

if any(p < 0.05 for p in [p_value_severe_moderate, p_value_severe_mild, p_value_severe_low]):
    print("There are significant differences in accident severity between different visibility conditions.")
else:
    print("There are no significant differences in accident severity between different visibility conditions.")
```
